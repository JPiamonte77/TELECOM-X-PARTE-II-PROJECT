```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix


# 1. Leer y preparar datos

df = pd.read_csv("TelecomX_Data_Procesado.csv")

# Eliminar filas con NaN en 'Churn'
df = df.dropna(subset=['Churn'])

# Separar variables predictoras y objetivo
X = df.drop(columns=['Churn'])
y = df['Churn']

# Dividir en entrenamiento y prueba (70/30) manteniendo proporciones
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# Modelo 1: Regresión Logística con normalización

# La Regresión Logística es sensible a la escala de las variables.
# La normalización (media=0, std=1) asegura que las variables con
# valores más grandes no dominen los coeficientes ni afecten la convergencia.
pipeline_logreg = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(max_iter=1000))
])

pipeline_logreg.fit(X_train, y_train)
y_pred_logreg = pipeline_logreg.predict(X_test)

print("Regresión Logística (con normalización) ")
print(classification_report(y_test, y_pred_logreg))
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_logreg))


# Modelo 2: Árbol de Decisión sin normalización


# Los árboles de decisión no dependen de la escala de las variables.
# Se basan en divisiones por umbrales, por lo que escalar no aporta mejora.
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)

print("\n Árbol de Decisión (sin normalización)")
print(classification_report(y_test, y_pred_tree))
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_tree))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

#  Normalización para el modelo sensible a la escala (Regresión Logística) ===
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Modelo 1: Regresión Logística (con normalización) ===
log_model = LogisticRegression(random_state=42, max_iter=1000)
log_model.fit(X_train_scaled, y_train)
y_pred_log = log_model.predict(X_test_scaled)

#  Modelo 2: Random Forest (sin normalización) ===
rf_model = RandomForestClassifier(random_state=42, n_estimators=200)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

#  Función para evaluar y mostrar métricas ===
def evaluar_modelo(nombre, y_true, y_pred):
    print(f" Resultados para {nombre}")
    print(f"Exactitud: {accuracy_score(y_true, y_pred):.4f}")
    print(f"Precisión: {precision_score(y_true, y_pred):.4f}")
    print(f"Recall: {recall_score(y_true, y_pred):.4f}")
    print(f"F1-score: {f1_score(y_true, y_pred):.4f}")
    print("\nReporte de Clasificación:\n", classification_report(y_true, y_pred))

    # Matriz de confusión
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["No Canceló", "Canceló"], yticklabels=["No Canceló", "Canceló"])
    plt.title(f"Matriz de Confusión - {nombre}")
    plt.ylabel("Valor Real")
    plt.xlabel("Predicción")
    plt.show()

#  Evaluar ambos modelos
evaluar_modelo("Regresión Logística", y_test, y_pred_log)
evaluar_modelo("Random Forest", y_test, y_pred_rf)
```

###Regresión Logística: análisis de coeficientes
Los coeficientes indican la influencia de cada variable en la probabilidad de que un cliente cancele (Churn=1).

Coeficiente positivo → aumenta la probabilidad de cancelación.

Coeficiente negativo → disminuye la probabilidad.

Para interpretarlos mejor, a menudo se usan los odds ratios (exp(coeficiente)).

```python
import numpy as np

# Suponiendo que pipeline_logreg ya está entrenado y X_train es DataFrame
coef = pipeline_logreg.named_steps['logreg'].coef_[0]
features = X_train.columns

# Crear DataFrame con coeficientes
df_coef = pd.DataFrame({'Variable': features, 'Coeficiente': coef})
df_coef['Odds_Ratio'] = np.exp(df_coef['Coeficiente'])

# Ordenar por valor absoluto del coeficiente para ver los más influyentes
df_coef['abs_coef'] = df_coef['Coeficiente'].abs()
df_coef = df_coef.sort_values(by='abs_coef', ascending=False)

print(df_coef[['Variable', 'Coeficiente', 'Odds_Ratio']])
```

### Random Forest: importancia de variables

Variables con mayor importancia son más relevantes para predecir.

###KNN y SVM
KNN no tiene coeficientes ni importancias directas; su predicción depende de la distancia a los vecinos. Para analizar qué variables influyen más, se suelen usar técnicas de reducción dimensional o análisis de sensibilidad.

SVM lineal tiene coeficientes como regresión logística (puedes revisar coeficientes de vectores de soporte). Para kernel SVM, la interpretación es menos directa.

###SVM Lineal: análisis de coeficientes
Cuando usas un SVM con kernel lineal, puedes acceder a los coeficientes que definen la frontera de decisión, similares a los coeficientes de una regresión logística. Estos coeficientes te indican qué variables influyen más en separar las clases.

```python
from sklearn.svm import LinearSVC
import numpy as np
import pandas as pd

# Entrenar SVM lineal (asegúrate de escalar variables antes)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm = LinearSVC(random_state=42, max_iter=5000)
svm.fit(X_train_scaled, y_train)

coef = svm.coef_[0]
features = X_train.columns

df_coef = pd.DataFrame({'Variable': features, 'Coeficiente': coef})
df_coef['abs_coef'] = df_coef['Coeficiente'].abs()
df_coef = df_coef.sort_values(by='abs_coef', ascending=False)

print(df_coef[['Variable', 'Coeficiente']])
```

### Redes Neuronales: pesos de las primeras capas
Los pesos en redes neuronales son más difíciles de interpretar directamente porque las variables se combinan no linealmente. Sin embargo, algunas técnicas de interpretación comunes incluyen:

SHAP (SHapley Additive exPlanations): calcula la importancia de cada variable en la predicción para cada instancia.

Permutación de importancia: evalúa el impacto de permutar valores de cada variable en la precisión del modelo.

###XGBoost y otros boosting
XGBoost y modelos de boosting basados en árboles calculan la importancia de las variables similar a Random Forest, basada en:

Ganancia: cuánto mejora el rendimiento el split con esa variable.

Frecuencia: cuántas veces se usa la variable para splits.

Cobertura: número de observaciones afectadas por esos splits.

```python
import xgboost as xgb
import matplotlib.pyplot as plt

xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train, y_train)

xgb.plot_importance(xgb_model, max_num_features=15)
plt.show()
```
